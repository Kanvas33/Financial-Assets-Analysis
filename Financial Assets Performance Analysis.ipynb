## Importing Libraries


```python
#Import the required libraries in the Python script

# For fetching stock data
import yfinance as yf
# For data manipulation and analysis
import pandas as pd
# For creating visualizations
import matplotlib.pyplot as plt
# For working with dates and times
from datetime import datetime, timedelta
# For numerical operations
import numpy as np
# For data visualization
import seaborn as sns
# For linear regression modeling
from sklearn.linear_model import LinearRegression
# For polynomial feature transformation
from sklearn.preprocessing import PolynomialFeatures
# For handling warnings or suppress warnings
import warnings
```

# Apple(AAPL) Stock price up to date


```python
# Define the stock symbol and set the end_date to today
stock_symbol = "AAPL"
start_date = "2020-01-01"
end_date = datetime.today().strftime('%Y-%m-%d')

# Fetch historical data and calculate daily returns
df = yf.download(stock_symbol, start=start_date, end=end_date)
df['Daily Returns'] = df['Adj Close'].pct_change()

# Create a plot for the stock
plt.figure(figsize=(12, 6))
line, = plt.plot(df.index, df['Adj Close'], label=f'{stock_symbol} Adjusted Close', color='blue')  # Store the line object
plt.title(f'{stock_symbol} Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# Get the latest closing price (today's price)
latest_price = df['Adj Close'].iloc[-1]

# Add a bullet point marker at the latest price
plt.scatter(df.index[-1], latest_price, color='black', marker='o', label='Today\'s Price')

# Annotate the latest price
plt.annotate(f'Today: {latest_price:.2f}', # Text to display
             xy=(df.index[-1], latest_price), # Position of the annotation (arrow tip)
             xytext=(df.index[-1], latest_price*0.95),  # Adjust the vertical position
             fontsize=12, # Font size of the text
             color='black') # Color of the text

# Display a timestamp outside the graphic grid (upper left)
timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')  # Get the current timestamp
plt.figtext(0.1, 0.9, f'Data as of: {timestamp}', fontsize=10, color='gray')

# Enable the grid
plt.grid(True)

# Show the plot
plt.legend()
plt.show()
```

##  BTC-USD with 200-Day and 50-Day EMA

```python
# Define the stock symbol and set the end_date to today
stock_symbol = "BTC-USD"
start_date = "2020-01-01"
end_date = datetime.today().strftime('%Y-%m-%d')

# Fetch historical data and calculate daily returns
df = yf.download(stock_symbol, start=start_date, end=end_date)
df['Daily Returns'] = df['Adj Close'].pct_change()

# Calculate the 200-day and 50-day Exponential Moving Averages (EMA)
df['EMA_200'] = df['Adj Close'].ewm(span=200, adjust=False).mean()
df['EMA_50'] = df['Adj Close'].ewm(span=50, adjust=False).mean()

# Create a plot for the stock
plt.figure(figsize=(12, 6))

# Plot Bitcoin's adjusted closing price
plt.plot(df.index, df['Adj Close'], label=f'{stock_symbol} Adjusted Close', color='purple')

# Plot the 200-day EMA
plt.plot(df.index, df['EMA_200'], label='200-Day EMA', color='red')

# Plot the 50-day EMA
plt.plot(df.index, df['EMA_50'], label='50-Day EMA', color='green')

plt.title(f'{stock_symbol} with 200-Day and 50-Day EMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# Get the latest closing price (today's price)
latest_price = df['Adj Close'].iloc[-1]

# Add a bullet point marker at the latest price
plt.scatter(df.index[-1], latest_price, color='black', marker='o', label='Today\'s Price')

# Annotate the latest price
plt.annotate(f'Today: {latest_price:.2f}', # Text to display
             xy=(df.index[-1], latest_price), # Position of the annotation (arrow tip)
             xytext=(df.index[-1], latest_price*0.95),  # Adjust the vertical position
             fontsize=12,  # Font size of the text
             color='black') # Color of the text

# Display a timestamp outside the graphic grid (upper left)
timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')  # Get the current timestamp
plt.figtext(0.1, 0.9, f'Data as of: {timestamp}', fontsize=10, color='gray')

# Enable the grid
plt.grid(True)

# Add explanatory text about 50 EMA and 200 EMA below the BTC graph
ema_text = """
50 EMA (Exponential Moving Average):

- The "50 EMA" stands for the 50-day Exponential Moving Average.
- It is a technical indicator that calculates the average closing price of a financial instrument (e.g., a stock) over the past 50 trading days.
- The Exponential Moving Average gives more weight to recent prices, making it more responsive to recent price movements compared to a simple moving average (SMA).
- Traders and analysts use the 50 EMA to identify trends and potential support/resistance levels in a stock's price. 
When the current price crosses above the 50 EMA, it can be seen as a bullish signal, indicating potential upward momentum. Conversely, when it crosses below, it can be seen as a bearish signal.

200 EMA (Exponential Moving Average):

- The "200 EMA" stands for the 200-day Exponential Moving Average.
- Similar to the 50 EMA, it calculates the average closing price of a financial instrument, but over a longer period of 200 trading days.
- The 200 EMA is often used to identify long-term trends and key support/resistance levels in a stock's price.
- When a stock's price is above the 200 EMA, it is considered to be in a long-term uptrend. Conversely, when the price is below the 200 EMA, it can indicate a long-term downtrend.
"""

# Create a text box with light gray background for the explanatory text
plt.figtext(0.1, -0.35, ema_text, fontsize=10, color='black', bbox=dict(facecolor='lightgray', alpha=0.2))

# Show the plot
plt.legend()
plt.show()


```

## BTC-USD Linear Regression model until 01/01/2028(4 years from today - 02/10/2023)

```python
# Define the stock symbol and set the end_date to today
stock_symbol = "BTC-USD"
start_date = "2014-01-01"
end_date = datetime.today().strftime('%Y-%m-%d')

# Fetch historical data and calculate daily returns
df = yf.download(stock_symbol, start=start_date, end=end_date)
df['Daily Returns'] = df['Adj Close'].pct_change()

# Calculate the 200-day and 50-day Exponential Moving Averages (EMA)
df['EMA_200'] = df['Adj Close'].ewm(span=200, adjust=False).mean()
df['EMA_50'] = df['Adj Close'].ewm(span=50, adjust=False).mean()

# Create a DataFrame for prediction
prediction_df = df[['Adj Close']].copy()
prediction_df['Numerical Date'] = pd.to_numeric(df.index)

# Define the number of future days for prediction - 4 years
future_days = 1565  # adjust this as needed

# Create a new DataFrame for future dates
future_dates = pd.date_range(start=df.index[-1], periods=future_days + 1, inclusive='right')
future_dates_numeric = pd.to_numeric(future_dates)

# Suppress the warning (C:\Users\jefff\anaconda3\lib\site-packages\sklearn\base.py:450: UserWarning: X does not have valid feature names, but LinearRegression was fitted with feature names)
warnings.filterwarnings("ignore", category=UserWarning)

# Create an instance of the LinearRegression model
model = LinearRegression()

# Fit a linear regression model to predict prices
model.fit(prediction_df[['Numerical Date']], prediction_df['Adj Close'])

# Predict future prices
future_prices = model.predict(future_dates_numeric.values.reshape(-1, 1))

# Create a DataFrame for future price predictions
future_df = pd.DataFrame({'Date': future_dates, 'Predicted Price': future_prices})
future_df.set_index('Date', inplace=True)

# Plot Bitcoin's adjusted closing price, 200-day EMA, and 50-day EMA
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Adj Close'], label=f'{stock_symbol} Adjusted Close', color='purple')
plt.plot(df.index, df['EMA_200'], label='200-Day EMA', color='red')
plt.plot(df.index, df['EMA_50'], label='50-Day EMA', color='green')

# Plot future price predictions
plt.plot(future_df.index, future_df['Predicted Price'], label='Predicted Price', linestyle='--', color='blue')

plt.title(f'{stock_symbol} with 200-Day and 50-Day EMA and Future Predictions')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# Get the latest closing price (today's price)
latest_price = df['Adj Close'].iloc[-1]

# Add a bullet point marker at the latest price
plt.scatter(df.index[-1], latest_price, color='black', marker='o', label='Today\'s Price')

# Annotate the latest price
plt.annotate(f'Today: {latest_price:.2f}', # Text to display
             xy=(df.index[-1], latest_price), # Position of the annotation (arrow tip)
             xytext=(df.index[-1], latest_price*0.95),  # Adjust the vertical position
             fontsize=12,  # Font size of the text
             color='black') # Color of the text

# Display a timestamp outside the graphic grid (upper left)
timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
plt.figtext(0.1, 0.9, f'Data as of: {timestamp}', fontsize=10, color='gray')

# Add explanatory text about the predicted price below the BTC graph
pred_price_text = """
Predicted Price:
It's important to note that linear regression models have limitations, especially when applied to financial markets like cryptocurrencies. 
Real-world price movements are influenced by numerous factors, including market sentiment, news events, 
and macroeconomic conditions, which may not be fully captured by a simple linear model. 
Therefore, while the predicted price provides an estimate, 
it should be interpreted with caution and in conjunction with other analyses when making investment decisions.
"""

# Create a text box with light gray background for the explanatory text
plt.figtext(0.125, -0.15, pred_price_text, fontsize=10, color='black', bbox=dict(facecolor='lightgray', alpha=0.2))

# Enable the grid
plt.grid(True)

# Show the plot
plt.legend(loc='upper left')
plt.show()

```

## Statistical analysis



```python
# Define a list of stock symbols
stocks = ["AAPL", "NVDA", "BTC-USD", "AMZN", "TSLA"]

# Define the date range
start_date = "2020-01-01"
end_date = datetime.today().strftime('%Y-%m-%d')  # This gets the current date in YYYY-MM-DD format


# Create an empty DataFrame to store stock data
stock_data = pd.DataFrame()

# Fetch historical data for each stock and concatenate it
for stock_symbol in stocks:
    stock = yf.download(stock_symbol, start=start_date, end=end_date)
    stock_data[stock_symbol] = stock['Adj Close']

# Calculate daily returns for each stock
returns = stock_data.pct_change()

# Calculate basic statistics for each stock
statistics = pd.DataFrame()
statistics['Mean Daily Return'] = returns.mean() * 100  # Convert to percentage
statistics['Volatility'] = returns.std() * 100  # Convert to percentage
statistics['Annualized Mean Return'] = statistics['Mean Daily Return'] * 252  # Assuming 252 trading days in a year
statistics['Annualized Volatility(in percentages):'] = statistics['Volatility'] * np.sqrt(252)  # Convert to percentage

# Display statistics with paragraph breaks
print("Basic Statistics (in percentages):")
print(statistics)

print("\n")

# Explanation of Annualized Volatility
explanation1 = """
Annualized volatility, also known as annualized standard deviation or annualized volatility, 
is a measure of the statistical 
dispersion or spread of returns for a financial asset (ex: stock) over a specific period, typically a year. 
It's commonly used in finance to assess the risk associated with an investment.
"""

print(explanation1)

# Separator line
print("\n" + "=" * 80 + "\n")

# Explanation of the correlation matrix
explanation2 = """ """

print(explanation2)

# Calculate correlation matrix
correlation_matrix = returns.corr() * 100  # Convert to percentage

# Display correlation matrix
print("\nCorrelation Matrix (in percentages):")


correlation_matrix
```

## Better visualization for the statistical analysis 
### Using Matplotlib.pyplot and Seaborn


```python
# Set a visually style with Seaborn
sns.set_style("whitegrid")
sns.set_palette("pastel")

# Create subplots for different visualizations
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))
fig.suptitle('Stock Market Analysis', fontsize=16)

# Plot Mean Daily Return
sns.barplot(x=statistics.index, y=statistics['Mean Daily Return'], ax=axes[0, 0], palette="Blues_d")
axes[0, 0].set_title('Mean Daily Return')
axes[0, 0].set_ylabel('Percentage')
axes[0, 0].set_xlabel('Stocks')

# Annotate Mean Daily Return
for bar in axes[0, 0].patches:
    axes[0, 0].annotate(f'{bar.get_height():.2f}%', (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                        ha='center', va='center', color='black', fontsize=10, xytext=(0, 7),
                        textcoords='offset points')

# Plot Volatility
sns.barplot(x=statistics.index, y=statistics['Volatility'], ax=axes[0, 1], palette="Greens_d")
axes[0, 1].set_title('Volatility')
axes[0, 1].set_ylabel('Percentage')
axes[0, 1].set_xlabel('Stocks')

# Annotate Volatility
for bar in axes[0, 1].patches:
    axes[0, 1].annotate(f'{bar.get_height():.2f}%', (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                        ha='center', va='center', color='black', fontsize=10, xytext=(0, 7),
                        textcoords='offset points')

# Plot Annualized Mean Return
sns.barplot(x=statistics.index, y=statistics['Annualized Mean Return'], ax=axes[1, 0], palette="Purples_d")
axes[1, 0].set_title('Annualized Mean Return')
axes[1, 0].set_ylabel('Percentage')
axes[1, 0].set_xlabel('Stocks')

# Annotate Annualized Mean Return
for bar in axes[1, 0].patches:
    axes[1, 0].annotate(f'{bar.get_height():.2f}%',
                        (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                        ha='center', va='center', color='black', fontsize=10, xytext=(0, 7),
                        textcoords='offset points')

# Plot Correlation Matrix
cax = sns.heatmap(correlation_matrix, ax=axes[1, 1], cmap='coolwarm', annot=True, fmt=".1f", cbar_kws={'label': 'Percentage'})
axes[1, 1].set_title('Correlation Matrix')


info_text = """
CORRELATION MATRIX INSIGHTS:

When you see a warm color (like 66.9) in the correlation matrix between AAPL and NVDA, 
it means these two tend to move in a similar direction. When one goes up, the other often goes up too.

On the other hand, when you see a cool color (like 32.5) in the correlation matrix between AAPL and BTC-USD, 
it means there isn't as strong of a relationship between them. They don't move in perfect harmony, 
and there's less consistency in how they move together.

In simpler terms, the correlation matrix helps you see if different things tend to move together (warm colors) 
or not (cool colors), which can be useful for making investment decisions or understanding how different assets 
in your portfolio behave in relation to each other.
"""

# Create a well-formatted text box, provides additional information or explanations about the data visualization.
## It allows to annotate plots with text in various parts of the figure, and you can control the text's size, color, and alignment.
info_box = axes[1, 1].annotate(info_text, xy=(0.00, -0.55), xycoords='axes fraction',
                              fontsize=10, color='black', ha='left',
                              bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.2))

# Display the plots
plt.show()
```
